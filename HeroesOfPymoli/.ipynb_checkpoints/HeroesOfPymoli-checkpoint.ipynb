
## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Total number of purchases by gender.

gender_group = file.groupby(['Gender'])

purchase_count = file["Gender"].value_counts()
average_purchase = gender_group["Price"].mean()
average_total = gender_group["Price"].sum()
purchase_total = average_total/gender_count

gender_analysis = pd.DataFrame({"Purchase Count":purchase_count, 
                                   "Average Purchase Price":average_purchase,
                                   "Total Purchase Value":average_total,
                                   "Average Purchase Total per Person":purchase_total})

pd.options.display.float_format = '${:,.2f}'.format

gender_analysis.round(2)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Create bins in which data will be held. Bins are <10, 10-14, 15-19, 20-24, 25-29, 30-34, 35-39 >39.

bins = [0,10,15,20,25,30,35,40, 45]
age_ranges = ["<10", "10-14","15-19", "20-24", "25-29", "30-34", "35-39", ">=40"]
# Cut purchase data and place the ages into bins
pd.cut(file["Age"], bins, labels=age_ranges)
file["Age Range"] = pd.cut(file["Age"], bins, labels= age_ranges)

# Categorize the existing players using the age bins
people["Age Range"] = pd.cut(people["Age"], bins, labels=age_ranges)

# Calculate the Numbers and Percentages by Age Group
age_demographics_totals = people["Age Range"].value_counts()
age_demographics_percents = age_demographics_totals / total * 100
age_demographics = pd.DataFrame({"Total Count": age_demographics_totals, "Percentage of Players": age_demographics_percents})

# Minor Data Munging
age_demographics = age_demographics.round(2)

pd.options.display.float_format = '{:,.2f}%'.format

# Display Age Demographics Table
age_demographics.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Bin the Purchasing Data
file["Age Ranges"] = pd.cut(file["Age"], bins, labels=age_ranges)

# Run basic calculations
age_purchase_total = file.groupby(["Age Ranges"]).sum()["Price"].rename("Total Purchase Value")
age_average = file.groupby(["Age Ranges"]).mean()["Price"].rename("Average Purchase Price")
age_counts = file.groupby(["Age Ranges"]).count()["Price"].rename("Purchase Count")

# Calculate Normalized Purchasing (Average Purchase Total per Person)
normalized_total = age_purchase_total / age_demographics["Total Count"]

# Convert to DataFrame
age_data = pd.DataFrame({"Purchase Count": age_counts, "Average Purchase Price": age_average, "Total Purchase Value": age_purchase_total, "Normalized Totals": normalized_total})

# Minor Data Munging
age_data["Average Purchase Price"] = age_data["Average Purchase Price"].map("${:,.2f}".format)
age_data["Total Purchase Value"] = age_data["Total Purchase Value"].map("${:,.2f}".format)
age_data ["Purchase Count"] = age_data["Purchase Count"].map("{:,}".format)
age_data["Avg Total Purchase per Person"] = age_data["Normalized Totals"].map("${:,.2f}".format)
age_data = age_data.loc[:, ["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]

# Display the Age Table
age_data
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Sort table to show the top spenders.
top_five_spenders = age_data.sort_values("Total Purchase Value", ascending=False)
top_five_spenders.head()
items_purchase_count_df = file.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")
items_average_price_df = file.groupby(["Item ID", "Item Name"]).mean()["Price"].rename("Average Purchase Price")
items_value_total_df = file.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")
items_purchased_df = pd.DataFrame({"Purchase Count":items_purchase_count_df,
                                   "Item Price":items_average_price_df,
                                   "Total Purchase Value":items_value_total_df,})

pd.options.display.float_format = '${:,.2f}'.format

items_purchased_df.head()
```
